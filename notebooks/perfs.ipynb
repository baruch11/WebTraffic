```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
board = pd.read_csv("../data/web-traffic-time-series-forecasting-publicleaderboard.zip")
```

```python
ax = board.Score.hist(bins=np.arange(30,80,.2),)
ax.set_xlim((30,50))
my_score = 39.53
plt.arrow(my_score, 10, 0, (1-10), color="r")
plt.text(my_score-2, 11, "my score", c="r")

baseline_score = 41.8
plt.arrow(baseline_score, 15, 0, (1-15), color="r")
plt.text(baseline_score-2, 16, "baseline score", c="r")

ax.set_title("kaggle webtraffic scores distribution\nin the leaderboard")
```

```python
f, ax = plt.subplots(figsize=(3,3))
res = ax.boxplot(board.Score)
ax.set_ylim(top=55)
plt.show()
```

```python
median = board.Score.quantile(.5)
q3 = board.Score.quantile(.75)
q1 = board.Score.quantile(.25)
outlier_limit = q3 + (q3 - q1) * 1.5
outlier_limit
```

```python
scores = board.Score[board.Score<outlier_limit]
```

```python
(scores<my_score).mean()
```
